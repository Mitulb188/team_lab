```python
import opensmile
import pandas as pd
import os
from tqdm import tqdm

# --- Configuration ---
asv_path = '/mount/studenten/arbeitsdaten-studenten1/team-lab-phonetics/2025/student_directories/AuFa/ASVSpoof19/LA'
output_dir = '/mount/studenten/arbeitsdaten-studenten1/team-lab-phonetics/2025/student_directories/AuFa/features_output'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# --- Data Loading Function ---
def load_asv_metadata(asv_path, partition='train'):
    """Loads the metadata for the ASVspoof 2019 dataset."""
    protocol_path = os.path.join(asv_path, f'ASVspoof2019_LA_cm_protocols/ASVspoof2019.LA.cm.{partition}.trn.txt')
    metadata = pd.read_csv(protocol_path, sep=' ', header=None)
    metadata.columns = ['speaker_id', 'audio_file_name', '_', 'system_id', 'label']
    audio_dir = os.path.join(asv_path, f'ASVspoof2019_LA_{partition}/flac')
    metadata['full_path'] = metadata['audio_file_name'].apply(lambda x: os.path.join(audio_dir, x + '.flac'))
    return metadata[['full_path', 'label']]

# --- Feature Extraction Setup ---
# 1. Initialize the openSMILE feature extractor for eGeMAPS
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPS,
    feature_level=opensmile.FeatureLevel.Functionals,
)

# --- Main Execution ---
try:
    # Load the dataset metadata
    df = load_asv_metadata(asv_path, 'train')
    print("✅ Successfully loaded metadata.")
    
    all_features = []
    print("\nStarting feature extraction with openSMILE...")

    # Loop through each audio file in the dataframe
    for index, row in tqdm(df.iterrows(), total=df.shape[0]):
        audio_file_path = row['full_path']
        try:
            # 2. Process the audio file
            features = smile.process_file(audio_file_path)
            
            # Add the file name and label for reference
            features['file_name'] = os.path.basename(audio_file_path)
            features['label'] = row['label']
            all_features.append(features)
        except Exception as e:
            print(f"\nCould not process file {audio_file_path}: {e}")

    # --- Save Results ---
    if all_features:
        # Combine the list of DataFrames into a single one
        features_df = pd.concat(all_features, ignore_index=True)

        # Reorder columns to have file_name and label first for clarity
        cols = ['file_name', 'label'] + [col for col in features_df.columns if col not in ['file_name', 'label']]
        features_df = features_df[cols]
        
        output_csv_path = os.path.join(output_dir, 'opensmile_eGeMAPS_features.csv')
        features_df.to_csv(output_csv_path, index=False)
        
        print(f"\n✅ Feature extraction complete!")
        print(f"Features for {len(features_df)} files saved to: {output_csv_path}")
        print("\nHere's a sample of your new features DataFrame:")
        print(features_df.head())
    else:
        print("\nNo features were extracted. Please check for errors.")

except FileNotFoundError:
    print("❌ Error: A file or directory was not found. Please double-check the 'asv_path'.")
```
